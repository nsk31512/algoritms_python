```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<b>Домашнее задание: оптимизировать x^2*(np.sin(0,5*x)^2+1)</b>
<ul>
<li>При каких значениях alpha оптимизация проходит успешно, а при каких "застревает"?</li></ul>

```python
def func(_x): return (_x**2)*(np.sin(0.5*_x)**2+1)
def grad(_x): return (2*_x)*(np.sin(0.5*_x)**2+1)+(_x**3)*np.cos(0.5*_x)**2
```

```python
x = np.linspace(-50, 50, 5000)
plt.plot(x, func(x))
plt.show()
```

```python
x = np.linspace(-50, 50, 5000)
plt.plot(x, grad(x))
plt.show()
```

```python
x = np.linspace(-20, 20, 500)
plt.plot(x, func(x))
plt.show()
```

```python
x = np.linspace(-10, 10, 500)
plt.plot(x, grad(x))
plt.show()
```

```python
x0=np.random.uniform(-10, 10)
alpha=0.035
while grad(x0) >= 0.01:
    print(f"x={x0}, f(x)={func(x0)}, f'(x)={grad(x0)}")
    x0-=alpha*grad(x0)
```

Оптимизация проходит успешно при alpha = 0,035 и x0 из интервала (-10, 10)
